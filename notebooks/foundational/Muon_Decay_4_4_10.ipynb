```python
# --- Colab Setup ---
import numpy as np
import matplotlib.pyplot as plt

# --- Scalar Bridge Parameters ---
epsilon_phi = 0.000354  # MeV (scalar energy scale from Appendix L)
E_mu_total = 105.658    # MeV (muon rest mass)
E_electron = 0.511      # MeV (electron rest mass)

# --- Energy Sweep ---
E_nu = np.linspace(0.1, E_mu_total - E_electron - 0.1, 500)
E_nue = E_mu_total - E_electron - E_nu

# --- Projected vector content (Bridge Equation) ---
q2_nu = E_nu / epsilon_phi
q2_nue = E_nue / epsilon_phi
q2_e = E_electron / epsilon_phi
q2_total = q2_nu + q2_nue + q2_e

# --- Plotting ---
plt.figure(figsize=(8, 5.5))
plt.plot(E_nu, q2_nu, label=r"$|q_{\nu_\mu}|^2$", color='orange')
plt.plot(E_nu, q2_nue, label=r"$|q_{\bar{\nu}_e}|^2$", color='red')
plt.plot(E_nu, q2_total, label=r"Total $|q_{\mathrm{projected}}|^2$", color='black', linewidth=2.5)

plt.xlabel(r"Muon Neutrino Energy $E_{\nu_\mu}$ (MeV)", fontsize=12)
plt.ylabel(r"Projected Vector Content $|q_{\mathrm{projected}}|^2$", fontsize=12)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("muon_scalar_projection_scan.png", dpi=300)
plt.show()
```
